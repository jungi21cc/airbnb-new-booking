```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```python
df_session_new = pd.read_csv("../data/merged_sessions.csv")
df_session = pd.read_csv("../data/sessions.csv")
df_train = pd.read_csv("../data/train_users_2.csv")
```

### 0. 준비

```python
df_train
```

```python
df_session_new.columns
```

```python
df_train_session = pd.merge(df_train, df_session_new, left_on = "id", right_on = "id")
```

```python
df_train_session["isDF"] = df_train_session["country_destination"].apply(lambda x : "NDF" if(x=="NDF") else "DF")
```

### 1. secs_elapsed

```python
fig, ax = plt.subplots(figsize=(5,5))
sns.distplot(df_train_session["secs_elapsed"], ax=ax)
```

```python
sns.boxplot(x = "isDF", y="secs_elapsed", data = df_train_session)
```

### 2. secs_sum 

```python
fig, ax = plt.subplots(figsize=(5,5))
sns.distplot(df_train_session["secs_sum"], ax=ax)
```

```python
sns.boxplot(x = "isDF", y="secs_sum", data = df_train_session)
```
